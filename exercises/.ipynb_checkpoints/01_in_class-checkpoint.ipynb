```python
supp = Chem.SDMolSupplier("../data/drugbank.sdf")
mols = []
count = 0
for mol in supp:
    if mol:
        mols.append(mol)
    else:
        count+=1
print(len(mols), count)
```

```python
m = mols[1]
type(m)
```

```python
rings = [Chem.GetSSSR(m) for m in mols]
```

```python
seaborn.histplot(rings)
```

```python
sum(rings)/len(rings)
```
